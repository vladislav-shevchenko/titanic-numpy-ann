```python
import numpy as np
import pandas as pd
from pandas import DataFrame as df
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
```

```python
titanic = pd.read_csv("train.csv")
```

```python
titanic
```

```python
imputer_med = SimpleImputer(missing_values=np.nan, strategy='median')
titanic[['Age']] = imputer_med.fit_transform(titanic[['Age']])
imputer_fr = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
titanic[['Embarked']] = imputer_fr.fit_transform(titanic[['Embarked']])
```

```python
print(titanic)
```

```python
x = titanic.iloc[:, [2, 4, 5, 6, 7, -1]].values

print(x)

y = titanic.iloc[:, 1].values

print(y)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
x[:, 1] = le.fit_transform(x[:, 1])
x[:, 5] = le.fit_transform(x[:, 5])
print(x)
```

```python
sc = StandardScaler()
x = sc.fit_transform(x)
print(x)
```

```python
w1 = np.random.rand(6,8)
w2 = np.random.rand(8,2)
reg = 0.003
step_size = 0.003
num_examples = x.shape[0]
print(w1)
print(w2)
print(x.shape)
```

```python
epoch = 40
for i in range(epoch):
    acc = 0
    for n in range(num_examples):
        h1 = np.maximum(0, np.dot(x[n], w1))
        out = np.dot(h1, w2)
        exp_out = np.exp(out)
        probs = exp_out / np.sum(exp_out)
        dscores = probs
        dscores[y[n]] -= 1
        dscores = np.matrix(dscores)
        h1 = np.matrix(h1)
        dW2 = np.dot(h1.T, dscores)
        dhidden = np.dot(dscores, w2.T)
        dhidden[h1 <= 0] = 0
        inp = np.matrix(x[n])
        dW = np.dot(inp.T, dhidden)
        dW2 += reg * w2
        dW += reg * w1
        w1 += -step_size * dW
        w2 += -step_size * dW2
        if np.argmax(out) == y[n]:
            acc += 1
            accuracy = acc / num_examples
    print("epoch =",i, "acc =", accuracy)
```

```python
titanic_test = pd.read_csv("test.csv")
titanic_test
```

```python
imputer_med = SimpleImputer(missing_values=np.nan, strategy='median')
titanic_test[['Age']] = imputer_med.fit_transform(titanic_test[['Age']])
imputer_fr = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
titanic_test[['Embarked']] = imputer_fr.fit_transform(titanic_test[['Embarked']])
```

```python
test = titanic_test.iloc[:, [1, 3, 4, 5, 6, -1]].values
print(test)
```

```python
test[:, 1] = le.fit_transform(test[:, 1])
test[:, 5] = le.fit_transform(test[:, 5])
print(test)
```

```python
test = sc.fit_transform(test)
n_e = test.shape[0]
print(test)
print(n_e)
```

```python
pred_list = []
for n in range(n_e):
    h1 = np.maximum(0, np.dot(test[n], w1))
    out = np.dot(h1, w2)
    if np.argmax(out) == 0:
        pred_list.append(0)
    else:
        pred_list.append(1)
```

```python
print(pred_list)
```

```python
data = {
    'PassengerId': range(892, 1310),
    'Survived': pred_list }

df = pd.DataFrame(data)

print(df)
```
